```python
import numpy as np
import pandas as pd
```

```python
samples_columns = pd.read_csv("mouse_VISp_2018-06-14_samples-columns.csv")
samples_columns
```

```python
samples_columns.columns
```

```python
GABA_cells = samples_columns.loc[samples_columns["class"] == "GABAergic"] # select GABAergic cells
GABA_cells
```

```python
# Seven uniqe subtype in this dataset
set(GABA_cells["subclass"])
```

```python
# seperate them
GABA_Lamp5_cells = GABA_cells.loc[GABA_cells["subclass"] == "Lamp5"]
GABA_Meis2_cells = GABA_cells.loc[GABA_cells["subclass"] == "Meis2"]
GABA_Pvalb_cells = GABA_cells.loc[GABA_cells["subclass"] == "Pvalb"]
GABA_Serpinf1_cells = GABA_cells.loc[GABA_cells["subclass"] == "Serpinf1"]
GABA_Sncg_cells = GABA_cells.loc[GABA_cells["subclass"] == "Sncg"]
GABA_Sst_cells = GABA_cells.loc[GABA_cells["subclass"] == "Sst"]
GABA_Vip_cells = GABA_cells.loc[GABA_cells["subclass"] == "Vip"]
```

```python
# read in count data (exon one for this demo)
exon_data = pd.read_csv("mouse_VISp_2018-06-14_exon-matrix.csv",
                        index_col = 0,)
exon_data
```

```python
# get each cell type's exon count matrix
GABA_Lamp5_exon = exon_data[GABA_Lamp5_cells['sample_name']]
GABA_Meis2_exon = exon_data[GABA_Meis2_cells['sample_name']]
GABA_Pvalb_exon = exon_data[GABA_Pvalb_cells['sample_name']]
GABA_Serpinf1_exon = exon_data[GABA_Serpinf1_cells['sample_name']]
GABA_Sncg_exon = exon_data[GABA_Sncg_cells['sample_name']]
GABA_Sst_exon = exon_data[GABA_Sst_cells['sample_name']]
GABA_Vip_exon = exon_data[GABA_Vip_cells['sample_name']]
```

```python
# save them as csv file for later analysis in R
GABA_Lamp5_exon.to_csv('GABA_Lamp5_exon.csv', index=True)
GABA_Meis2_exon.to_csv('GABA_Meis2_exon.csv', index=True)
GABA_Pvalb_exon.to_csv('GABA_Pvalb_exon.csv', index=True)
GABA_Serpinf1_exon.to_csv('GABA_Serpinf1_exon.csv', index=True)
GABA_Sncg_exon.to_csv('GABA_Sncg_exon.csv', index=True)
GABA_Sst_exon.to_csv('GABA_Sst_exon.csv', index=True)
GABA_Vip_exon.to_csv('GABA_Vip_exon.csv', index=True)
```
