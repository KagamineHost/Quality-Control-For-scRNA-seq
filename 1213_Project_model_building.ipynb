```python
## 7 cell types verrsion
pca_dat = pd.read_csv('pca_cellembeddings.csv', sep = ' ')
pca_lab = pd.read_csv('1211_celltype_label.csv', sep=',') 
```

```python
## 7 cell types version
lab_fac = ["Lamp5", "Meis2", "Pvalb", "Serpinf1", "Sncg", "Sst", "Vip"]
```

```python
pca_dat
pca_dat_sub = pca_dat.iloc[:,range(30)]
pca_dat_sub
```

```python
# Split into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(pca_dat_sub, pca_lab, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lrmodel = LogisticRegression()
```

```python
lrmodel_p = LogisticRegression(penalty='l1', solver='liblinear')
```

```python
lrmodel_balance = LogisticRegression(class_weight = 'balanced')
```

```python
# fit the logistic regression model
lrmodel.fit(x_train, y_train)
```

```python
lrmodel_p.fit(x_train, y_train)
```

```python
lrmodel_balance.fit(x_train, y_train)
```

```python
# predcit
lrmodel.score(x_test,y_test)
y_pre = lrmodel.predict(x_test)
y_pre_proba = lrmodel.predict_proba(x_test)
```

```python
lrmodel_p.score(x_test,y_test)
y_pre_p = lrmodel.predict(x_test)
y_pre_proba_p = lrmodel.predict_proba(x_test)
```

```python
lrmodel_balance.score(x_test,y_test)
y_pre_balance = lrmodel_balance.predict(x_test)
y_pre_proba_balance = lrmodel_balance.predict_proba(x_test)
```

```python
lrmodel.score(x_test,y_test)
```

```python
lrmodel_balance.score(x_test,y_test)
```

```python
import sklearn.metrics as metrics

fpr = {}
tpr = {}
thresh ={}
auc = {}
plt.figure(figsize = (10,7))
for i in range(7):    
    fpr[i], tpr[i], thresh[i] = metrics.roc_curve(y_test, y_pre_proba_p[:,i], pos_label=lab_fac[i])
    auc[i] = metrics.auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label = 'AUC %0.1f'% i+'= %0.7f' % auc[i])
plt.title('Penalized logistic regression ROC')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
```
